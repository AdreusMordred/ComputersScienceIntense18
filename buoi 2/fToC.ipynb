# Import thЖ° viб»Үn

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

# Import dб»Ҝ liб»Үu

```python
df = pd.read_csv("temperatures.csv")
df
```

# Chuбә©n hГіa

```python
f_input = df["Farenheit"].to_numpy()
c_input = df["Celcius"].to_numpy()
type(f_input), type(c_input)
```

# Train mГҙ hГ¬nh

```python
model = keras.Sequential([
    keras.layers.Dense(units=10, input_shape=[1]),  #Layer Dense 1 vб»ӣi 10 neuron -> input shape lГ  1 vГ¬ chб»ү cГі giГЎ trб»Ӣ C
    keras.layers.Dense(units=1) #Layer Dense 2 vб»ӣi 1 neuron -> trбәЈ vб»Ғ giГЎ trб»Ӣ C
])

model.compile(
    loss = "mean_squared_error", #HГ m mбәҘt mГЎt lГ  mean_squared_error
    optimizer = "adam"
)

model.summary()
```

```python
# Train mГҙ hГ¬nh
model.fit(f_input, c_input, epochs=100) #Epochs lГ  sб»‘ lбә§n lбә·p qua toГ n bб»ҷ dб»Ҝ liб»Үu huбәҘn luyб»Үn
```

# Test mГҙ hГ¬nh

```python
f_user_input = np.array([90])

# Thб»ұc tбәҝ
c_output_real = round((f_user_input[0] - 32) / 1.8, 2)
print(f"GiГЎ trб»Ӣ C thб»ұc tбәҝ: {c_output_real}")

# Dб»ұ Д‘oГЎn
c_output_pred = model.predict(f_user_input)
print(c_output_pred)
print(f"GiГЎ trб»Ӣ C dб»ұ Д‘oГЎn: {c_output_pred[0][0]:.2f}")
```
